## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
numeric_vars=['BorrowerAPR', 'LoanOriginalAmount','MonthlyLoanPayment'] 
category_vars=['Term','LoanStatus','EmploymentStatus']
```

#### correlation between variables

```python
# correlation(matrices) plot
plt.figure(figsize = [8,5])
sb.heatmap(sub_loan[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

```python
g=sb.PairGrid(data=sub_loan,vars=['BorrowerAPR','LoanOriginalAmount','MonthlyLoanPayment']);
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter,alpha=0.2)
```

#### Correlation between Monthly Loan Payment and Loan Original Amount

```python
sub_loan['MonthlyLoanPayment'].describe()
```

```python
sub_loan['LoanOriginalAmount'].describe()
```

```python
# Specify bin edges 
bins_x = np.arange(0, 2251+150,150)
bins_y = np.arange(1000, 35000+900,900)

plt.hist2d(data = sub_loan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', cmin=0.5, cmap='viridis_r', bins = [bins_x, bins_y])
plt.colorbar()
plt.xlabel('MonthlyLoanPayment')
plt.ylabel('LoanOriginalAmount');
```

There is a strong correlation between both variables as LoanOriginalAmount increases, MonthlyLoanPayment increases.
most likely monthly payment 1000$ and few how pay more than that.  

#### EmploymentStatus VS Term

```python

sb.countplot(data = sub_loan, x = 'EmploymentStatus', hue = 'Term',palette = 'Blues');
plt.xticks(rotation=90)
```

There is only 36 and 60 months loans for employed borrowers. although, retired ,part-time employed borrowers only have 36 months loan.

#### Correlation between BorrowerARP and original loan amount 

we want to see how they are related 

```python
sb.regplot(data=sub_loan,x='LoanOriginalAmount',y='BorrowerAPR',scatter_kws={'alpha':1/5})
plt.xlabel('LoanOriginalAmount')
plt.ylabel('BorrowerAPR')
```

here there is a negative correlation between the two variables. when loan amount increases the borrowerARP decreases
