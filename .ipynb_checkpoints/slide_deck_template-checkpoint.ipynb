## Dataset Overview

> The data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
loan=pd.read_csv('prosperLoanData.csv')
col=['Term','BorrowerAPR','BorrowerRate','LoanOriginalAmount','LoanStatus','ProsperRating (Alpha)','ProsperRating (numeric)','MonthlyLoanPayment','EmploymentStatus']
sub_loan=loan[col]
```

## Distribution of Borrower APR 

>As we see BorrowerAPR distribution is mulitmodel there are several peaks small peak at 0.1 and large one at 0.2 another 2 peaks at 0.3 and 0.35. 

```python
plt.figure(figsize=[10, 6])
bins=np.arange(sub_loan['BorrowerAPR'].min(),sub_loan['BorrowerAPR'].max()+0.01,0.01)
plt.hist(data = sub_loan, x = 'BorrowerAPR',bins=bins);
plt.xlabel('Borrower APR(%)');
plt.ylabel('Count');
plt.title('Distribution of Borrower APR');

```

## Correlation between BorrowerARP and original loan amount 

> The relationship between loan amount and Borrower ARP is negative. There is strong effect on Borrower ARP by Loan amount which when loan amount increases the Borrower ARP decreases.

```python
plt.figure(figsize=(10,6))
sb.regplot(data=sub_loan,x='LoanOriginalAmount',y='BorrowerAPR',scatter_kws={'alpha':1/5})
plt.xlabel('LoanOriginalAmount($)')
plt.ylabel('BorrowerAPR(%)')
plt.title('BorrowerARP and LoanOriginalAmount')
```

## Correlation between Monthly Loan Payment and Loan Original Amount


>There is a strong correlation between both variables as LoanOriginalAmount increases, MonthlyLoanPayment increases.
most likely monthly payment $1000 and few how pay more than that.  

```python
# Specify bin edges 
bins_x = np.arange(0, 2251+150,150)
bins_y = np.arange(1000, 35000+900,900)

# plotting the data
plt.figure(figsize=(10,6))
plt.hist2d(data = sub_loan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', cmin=0.5, cmap='viridis_r', bins = [bins_x, bins_y])
plt.colorbar()
plt.xlabel('MonthlyLoanPayment($)')
plt.ylabel('LoanOriginalAmount($)');
plt.title('Monthly Loan Payment VS Loan Original Amount')
```

## How Term affect MonthlyLoanPayment and LoanOriginalAmount


>It seems that term affect on both two variables and when loan amount increases so as term variable increse monthly payment decrease. 

```python


plt.figure(figsize=(10,8))
g = sb.FacetGrid(data = sub_loan, col = 'Term',size = 2.5,
                margin_titles = True);
g.map(plt.scatter, 'MonthlyLoanPayment', 'LoanOriginalAmount',alpha=0.7,);
g.set_xlabels('Monthly Loan Payment($)');
g.set_ylabels('Loan Original Amount($)');
plt.suptitle('Term Effect on Relationship between Monthly payment and Loan Amount');
plt.subplots_adjust(top=0.70);
```
